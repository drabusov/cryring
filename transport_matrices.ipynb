```python
def matrix_elem(index):

    r11 = float(a.R11[index])
    r12 = float(a.R12[index])
    r21 = float(a.R21[index])
    r22 = float(a.R22[index])
    
    return [[r11, r12],[r21, r22]]


def transport(initial, final, arr_matrices):

    m = arr_matrices[initial]
    for i in range(initial+1, final):
        m = np.matmul(m,arr_matrices[i])
    return m
```

```python
matrices =[matrix_elem(i) for i in range(1, len(a.R11))]
```
